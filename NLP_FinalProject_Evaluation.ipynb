#Inference for Bloomz


```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load model and tokenizer
model_directory = "sprab4/bloomz_fine_tuned_model"
model = AutoModelForCausalLM.from_pretrained(model_directory)
tokenizer = AutoTokenizer.from_pretrained(model_directory)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

while True:
    prompt = input("Enter a question in Haitian Creole (or type 'exit' to quit): ")
    if prompt.lower() == "exit":
        print("Exiting...")
        break

    formatted_prompt = f"{prompt}\nRepons:"  # Adding 'Repons:' signals the start of the response

    # Tokenize the input
    inputs = tokenizer(
        formatted_prompt,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=512
    ).to(device)

    # Generate response
    outputs = model.generate(
        input_ids=inputs["input_ids"],
        max_length=128,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True
    )

    # Decode and print the response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"{response}\n")
```

#Inference for mt5

```python
from transformers import MT5ForConditionalGeneration, MT5Tokenizer
import torch

# Replace this with your Hugging Face model repository name
model_name = "sprab4/mt5_fine_tuned_model"

# Load the model and tokenizer from Hugging Face
model = MT5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = MT5Tokenizer.from_pretrained(model_name)

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

while True:
    # Get the Haitian Creole instruction from the user
    user_instruction = input("Enter your instruction in Haitian Creole (or type 'exit' to quit): ")

    if user_instruction.lower() == "exit":
        print("Exiting...")
        break

    # If you have no separate "input" context (other than the instruction), you can leave it blank
    prompt = f"""Below is an instruction that describes a task. Write a response that appropriately completes the request.

### Instruction:
{user_instruction}

### Input:


### Response:
"""

    # Encode the prompt
    input_ids = tokenizer(
        prompt,
        return_tensors="pt",
        add_special_tokens=True,
        padding=True,
        truncation=True
    ).input_ids.to(device)

    # Generate the output
    output_ids = model.generate(
        input_ids=input_ids,
        max_length=100,
        num_return_sequences=1,
        do_sample=True, # picks tokens randomly by weighing the probability
        top_k=50, # Increases diversity (only considers k highest probable tokens)
        top_p=0.9, # Selects a pool of tokens where their combined probability equals 0.9
        temperature = 0.7
    )[0]

    # Decode the output
    output_text = tokenizer.decode(output_ids, skip_special_tokens=True)
    print(f"Generated text: {output_text}")
```

#Evaluating the Models using BERTScore

```python
#Evaluating for Bloomz Model
from transformers import AutoModelForCausalLM, AutoTokenizer
from bert_score import score
import torch

model_directory = "sprab4/bloomz_fine_tuned_model"
model = AutoModelForCausalLM.from_pretrained(model_directory)
tokenizer = AutoTokenizer.from_pretrained(model_directory)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Move model to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

prompts = [
    "Solèy la ap kouche ak yon bèl solèy kouche vizib nan orizon an, dekri li nan de fraz",
    "Jenere yon lis 10 atik ki ka itilize kòm yon kòmanse konvèsasyon",
    "Ajoute senk adjektif pou dekri son yon kaskad dlo"
]
references = [
    "Syèl la limen ak koulè zoranj, woz ak koulè wouj violèt kòm solèy la plonje anba orizon an, pentire yon twal mayifik nan syèl la. Ekla cho solèy kouche a benyen tout bagay nan chemen li yo ak yon limyè an lò, kreye yon vizyon trankilite ak fasinan pou wè",
    "1. Ki sa ki pastan ou pi renmen oswa al pran plezi? 2. Ki bèt ou pi idantifye ak poukisa? 3. Si ou ta ka vizite nenpòt kote nan mond lan, ki kote li ta ye e poukisa? 4. Ki liv oswa fim ki te gen pi gwo enpak sou ou, e poukisa? 5. Ki akonplisman ou pi fyè? 6. Ki bagay ki pi espontane ou janm fè? 7. Ki moman ki te pi kontan nan lavi ou jiskaprezan? 8. Èske ou kwè nan sò, oswa ou panse nou kontwole pwòp desten nou an? 9. Si ou ta ka reviv nenpòt moman nan lavi ou, kisa li ta ye? 10. Ki sa ou renmen plis nan travay ou, oswa ki sa ki ta travay rèv ou?",
    "1. Loraj 2. Gwonde 3. Trankil 4. Kraze 5. Kalman"
]

def generate_predictions(model, tokenizer, prompts, task_marker="Repons:"):
    predictions = []
    for prompt in prompts:
        formatted_prompt = f"{prompt}\n{task_marker}"

        # Tokenize the input
        inputs = tokenizer(
            formatted_prompt,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=512
        ).to(device)

        # Generate response
        outputs = model.generate(
            input_ids=inputs["input_ids"],
            max_length=128,
            num_beams=5,
            no_repeat_ngram_size=2,
            early_stopping=True
        )

        # Decode the response
        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
        clean_response = response.replace(task_marker, "").strip()
        predictions.append(clean_response)
    return predictions

print("Generating predictions...")
bloomz_predictions = generate_predictions(model, tokenizer, prompts, task_marker="Repons:")

print("Evaluating with BERTScore...")
P, R, F1 = score(bloomz_predictions, references, lang="ht", verbose=True)

print("\nEvaluation Results:")
for i, (prompt, ref, pred) in enumerate(zip(prompts, references, bloomz_predictions)):
    print(f"Prompt: {prompt}")
    print(f"Reference: {ref}")
    print(f"Prediction: {pred}")
    print(f"BERTScore Precision: {P[i].item():.4f}")
    print(f"BERTScore Recall: {R[i].item():.4f}")
    print(f"BERTScore F1: {F1[i].item():.4f}\n")
```
